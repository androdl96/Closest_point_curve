```python
import geojson
import numpy as np
from mmap import MAP_ANONYMOUS
import folium
#from shapely.geometry import LineString, Point
#from shapely.ops import transform
#import pyproj
#from math import ceil
import random
import matplotlib.pyplot as plt
```

```python
with open('Closest_point_curve/lijnsecties-1.json', 'r') as f:
    data = geojson.load(f)
```

```python
trajectories_size=[len(np.array(data['features'][i]['geometry']['coordinates'])) for i in range(len(data['features']))]
print(np.argmax(trajectories_size),np.max(trajectories_size))
```

```python
line_0 = np.array(data['features'][360]['geometry']['coordinates'])
line_0[:,[0,1]]=line_0[:,[1,0]]
map_center=line_0[400,:]
mapa = folium.Map(location=map_center, zoom_start=13)
folium.PolyLine(locations=line_0, color='red', weight=3).add_to(mapa)
line_1 = np.array(data['features'][445]['geometry']['coordinates'])
line_1[:,[0,1]]=line_1[:,[1,0]]
folium.PolyLine(locations=line_1, color='blue', weight=3).add_to(mapa)
line_2 = np.array(data['features'][513]['geometry']['coordinates'])
line_2[:,[0,1]]=line_2[:,[1,0]]
folium.PolyLine(locations=line_2, color='green', weight=3).add_to(mapa)
#line_3 = np.array(data['features'][419]['geometry']['coordinates'])
line_3 = np.array(data['features'][239]['geometry']['coordinates'])
line_3[:,[0,1]]=line_3[:,[1,0]]
folium.PolyLine(locations=line_3, color='black', weight=5).add_to(mapa)
mapa
```

```python
line=np.concatenate((line_2[:-1,:],line_1[:-1,:],line_0[:-1,:],line_3),axis=0)
print(len(line))
map_center=line_0[500,:]
map_line = folium.Map(location=map_center, zoom_start=11)
folium.PolyLine(locations=line, color='red', weight=3).add_to(map_line)
map_line
```

```python
np.savetxt('line_BEL_extended.txt', line, delimiter=',', comments='',fmt='%.7f')
print("Datos guardados en 'line_BEL_extended.txt'")
```
